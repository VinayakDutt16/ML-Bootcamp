```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/HP/Downloads/click.csv")
```

```python
df.head()
```

```python
df = df.drop(columns = "Timestamp",axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['Ad Topic Line'] = lb.fit_transform(df['Ad Topic Line'])
df['City'] = lb.fit_transform(df['City'])
df['Gender'] = lb.fit_transform(df['Gender'])
df['Country'] = lb.fit_transform(df['Country'])
```

```python
df.head()
```

```python
x = df.drop(columns = "Clicked on Ad")
y = df["Clicked on Ad"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
np.round(x_train.describe(),1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn = MinMaxScaler()
```

```python
x_train_mn = mn.fit_transform(x_train)
x_test_mn = mn.transform(x_test)
```

```python
x_train_new = pd.DataFrame(x_train_mn, columns = x_train.columns)
x_test_new = pd.DataFrame(x_test_mn, columns = x_test.columns)
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/HP/Downloads/click.csv")

df['Timestamp'] = pd.to_datetime(df['Timestamp'])


df['Month'] = df['Timestamp'].dt.month


```

```python
df.head()
```

```python
max_clicks_per_month = df.groupby('Month')['Clicked on Ad'].max()

# Displaying the result
print(max_clicks_per_month)
```

```python
max_clicks_month = max_clicks_per_month.idxmax()
print(f"The month with the maximum clicks is: {max_clicks_month}")
```
